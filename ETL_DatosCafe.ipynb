Importar Librerías

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import requests
```

Carga de Datos

```python
url='https://raw.githubusercontent.com/JP-UC/CoffeFiles/master/Exportaciones_Cafe_Federacion.xlsx'
_fexports=requests.get(url)
dest = 'tmp.xlsx'
with open(dest,'wb') as file:
  file.write(_fexports.content)

df_hist_expo=pd.read_excel(dest,sheet_name="Hist_Exp")
df_hist_expo=df_hist_expo.rename(columns={"País destino":"Pais_Destino"
                                                            ,"Sacos de 70 kg. equivalente real Exportados":"Sacos_70kg_Exportados"
                                                            ,"Sacos de 60 Kg. Exportados":"Sacos_60kg_Exportados"
                                                            ,"Valor Factura (USD)*":"Valor_USD"
                                                            ,"Nombre exportador":"Exportador"})
#df_hist_exportaciones.head()
#df_hist_expo.describe()

df_hist_expo_tipo=pd.read_excel(dest,sheet_name="Hist_Exp_Tipo_Cafe")
df_hist_expo_tipo=df_hist_expo_tipo.rename(columns={"Puerto de embarque":"Puerto_Embarque"
                                                            ,"Tipo de café":"Tipo_Cafe"
                                                            ,"Sacos de 70 kg. equivalente real Exportados":"Sacos_70kg_Exportados"
                                                            ,"Sacos de 60 Kg. Exportados":"Sacos_60kg_Exportados"
                                                            ,"Valor provisional de la Exportación (USD) *":"Valor_USD"
                                                            }
                                           )

df_hist_expo_tipo.head()
#df_hist_exportaciones.describe()
```
